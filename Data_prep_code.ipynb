```python
import os
import shutil
import pandas as pd

# Define the path to the main image directory
  # Replace with the actual path to your "images" folder
output_directory = "Your path of where you would want the new images" 
image_directory = "Your path to the images"
# Create the output directory if it doesnt exist
os.makedirs(output_directory, exist_ok=True)

# Initialize variables
filenames = []
labels = []
image_number = 1  # Start numbering images from 1

# Sort folders and iterate through each folder in numerical order
for folder_name in sorted(os.listdir(image_directory)):
    label_folder = os.path.join(image_directory, folder_name)
    if os.path.isdir(label_folder):  # Ensure its a directory
        # Use the full folder name as the label (e.g., "001.Black_footed_Albatross")
        label = folder_name
        
        # Sort image files within the folder to ensure consistent order
        for image_file in sorted(os.listdir(label_folder)):
            # Only include image files with common extensions
            if image_file.endswith(('.jpg', '.jpeg', '.png')):
                # Define the new filename with the sequential image number
                new_filename = f'Image_{image_number}.jpg'
                
                # Print each filename and label to verify order (temporary check)
                print(f"{new_filename} - {label}")

                # Copy the image to the output directory with the new name
                source = os.path.join(label_folder, image_file)
                destination = os.path.join(output_directory, new_filename)
                shutil.copy2(source, destination)
                
                # Append to the lists for DataFrame creation
                filenames.append(new_filename)
                labels.append(label)
                
                # Increment the image number for the next file
                image_number += 1

# Create a DataFrame with the new filenames and corresponding labels
df = pd.DataFrame({
    'filename': filenames,
    'label': labels
})

# Save the DataFrame to a CSV file
df.to_csv('warbler_image_labels.csv', index=False)

# Display the first few rows of the DataFrame for verification
print(df.head())
```
